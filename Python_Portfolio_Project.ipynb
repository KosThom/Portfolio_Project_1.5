```python
# Let us order our data by gross revenue

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# We will also do some slicing and see the top 10 movies by gross revenue

df.sort_values(by='gross', ascending = False)[:10]
```

```python
# Now looking at the top 10 film companies by gross revenue (2007-2016)

CompanyGrossSum = df.groupby('company')[['gross']].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending=False)[:10]

CompanyGrossSumSorted
```

```python
# Let us create a simple visualization (horizontal bar chart) for the aforementioned data

Gross_Bar_Chart = CompanyGrossSumSorted.plot.barh(color='blue')

plt.title('Top-10 Film Companies by gross revenue (2007-2016)')

plt.xlabel('gross revenue*10bn $')
```

```python
# Now we would like to see the total gross revenue by year

YearGrossSum = df.groupby('year')[['gross']].sum()

YearGrossSumSorted = YearGrossSum.sort_values('year', ascending=False)

YearGrossSumSorted
```

```python
# Now to create a line chart
Gross_line_Chart = YearGrossSumSorted.plot.line(color='green')

plt.title('Total gross revenue of film industry (2007-2016')

plt.ylabel('gross revenue in bn $')
```

```python
# It seems that the 30-year period examined here there were five drops in gross revenue for the film industry,
# namely in the years 1991,1998,2006,2011 and 2014. Nevertheless, in 1986 the gross revenue was $2.6 billion ,
# whereas in 2016 almost $9 billion.

# We would now like to check the hypothesis that there is a positive correlation between
# the budget of a movie and the gross revenue it makes. To check that we are going to create
# a scatter plot


plt.scatter(x=df['gross'], y=df['budget'], alpha = 0.35)

plt.title('Budget vs Gross revenue')

plt.xlabel('Gross revenue in bn $')

plt.ylabel('Budget*100m $')

plt.show()
```

```python
# In order to check if a positive correlation between budget and gross revenue exists
# we are going to create a regression plot using seaborn

sns.regplot(x='gross', y='budget', data=df, scatter_kws={"color":'black'}, line_kws={'color':'red'})

plt.title('Budget vs Gross revenue with regression line')

plt.xlabel('Gross revenue in bn $')

plt.ylabel('Budget*100m $')
```

```python
# We can clearly see that there is a positive linear relationship between the budget of a 
# film and the gross revenue it makes. We could also measure that by creating a 
# correlation matrix. In fact we will check if there is also a correlation between the budget
# and the score. That means that we only need these three numeric
# columns for the correlation matrix we are going to build (we do not need runtime and year)

corr_all_df=df.corr(method = 'pearson')

corr_three_df=corr_all_df.drop(['runtime','year']).drop(['runtime','year'],axis=1)

corr_three_df
```

```python
# It is obvious that there is a positive correlation between budget and gross revenue
# (correlation coefficient = 0.683), exactly as the scatterplot with regression line above showed us.
# On the contrary, we can see that there is almost no correlation at all between the budget of a film and its
# score (correlation coefficient = 0.02). First let us create a heatmap from the correlation matrix in order to have a better visualization

sns.heatmap(corr_three_df, annot= True, cmap="YlGnBu")

plt.title("Correlation matrix for three values")
```

```python
# Now to create the scatterplot with the regression line (budget vs score)

sns.regplot(x='score', y='budget', data=df, scatter_kws={"color":'red'}, line_kws={'color':'blue'})

plt.title('Budget vs Score with regression line')

plt.xlabel('Score in imdb (0-10)')

plt.ylabel('Budget*100m $')
```

```python
# It is clear-just as the correlation matrix showed us-that higher budget
# does not mean that the audience is going to give the movie a high score
```
